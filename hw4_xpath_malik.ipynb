## MAIL.RU

```python
import requests
import json
import pandas as pd
import numpy as np
from lxml import html
from pprint import pprint
from pymongo import MongoClient

the_news=[]

main_link = 'https://news.mail.ru'
header = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.132 Safari/537.36'}
response = requests.get(main_link, headers=header).text
tree = html.fromstring(response)

news_block = tree.xpath('//div[@class="newsitem newsitem_height_fixed js-ago-wrapper"]')


for i in news_block:
    the_list={}
    news = (i.xpath('.//span[@class="newsitem__title-inner"]/text()')[0])
    link = (main_link + i.xpath('.//@href')[0])
    date = (i.xpath('.//@datetime')[0][:10])
    source = (i.xpath('.//span[@class="newsitem__param"]/text()')[0])
    
    the_list['news_1']= news
    the_list['link_1']= link
    the_list['date']= date
    the_list['source']= source
    
    the_news.append(the_list)
    
json = the_news
json

df = pd.DataFrame(json) 
df.head()

dict=df.to_dict('records')#!!! Craetes a JSON 
df.head()
```

```python
try: 
    client = MongoClient('localhost', 27017)
    print("Connected successfully!!!") 
except:   
    print("Could not connect to MongoDB") 
    
db = client['MAIL_RU']
collection = db['collection']


try: 
    collection.insert_many(json)
    print("Loaded Successfully") 
except Exception as e:   
    print(e)
```

## YANDEX.RU

```python
the_news=[]

main_link = 'https://yandex.ru/news'

header = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.132 Safari/537.36'}

response = requests.get(main_link, headers=header).text
tree = html.fromstring(response)

news_block = tree.xpath('//div[@class="story story_view_short story_notags"]')

    
for i in news_block:
    the_list={}
    date = i.xpath(".//div[@class='story__date']/text()")[0].split(' ')[-1]          
    news = (i.xpath('.//text()')[0])
    link = ('https://yandex.ru' + (i.xpath('.//@href')[0]))
    date = date 
    source = (i.xpath('.//div[@class="story__date"]/text()')[0]).replace(date,'')
    
    the_list['news_1']= news
    the_list['link_1']= link
    the_list['date']= date
    the_list['source']= source
    
    the_news.append(the_list)
    
json = the_news
json

df = pd.DataFrame(json) 
df.head()
```
